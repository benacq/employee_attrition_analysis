## ANALYSIS OBSERVATIONS

##### After the analysis there were some key observations were made which could be a contributing factor to the employee attrition problem

- Some of the employees left because they are not satisfied with the work for some reason, which could be low salary, promotional problem etc.

- Some employees also left for no traceable reason, from the analysis some of the employees that left had very good satisfaction level, high salary and very good evaluation yet they still left.

- Also some employees that left had as high as 300 montly working hours, probably they left because they felt they were being overworked and under paid since most of the employees that left had low salaries

- Head of departments, especially for the sales, technical and support departments should ensure that their employees are well monitered because most of the employees that left were coming from these departments respectively.

- Also the companies management board should make sure employees are not overworked, montly working hours should be kept below 230

- Also management should make sure their employees are well paid for the number of hours they work, if employees feel they are overworked and underpaid, they are very likely leave for a better job.


## BUILDIING MODEL

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
employees['salary'] = le.fit_transform(employees['salary'])
employees['status'] = le.fit_transform(employees['status'])
employees['salary']
```

```python
employees['status']
```

```python
employees.drop('dept', inplace= True, axis= 1)
```

```python
emp_exist.head()
```

```python
employees.head()
```

```python
X= employees[['satisfaction_level', 'last_evaluation', 'number_project',
       'average_montly_hours', 'time_spend_company', 'Work_accident',
       'promotion_last_5years', 'salary']].values
Y = employees[['status']].values
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = .25)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

SVC = SVC()
RF = RandomForestClassifier()
KNN = KNeighborsClassifier()
DTC = DecisionTreeClassifier()
```

```python
SVC.fit(X_train, y_train)
RF.fit(X_train, y_train)
KNN.fit(X_train, y_train)
DTC.fit(X_train, y_train)
```

```python
SVC_prediction = SVC.predict(X_test)
DTC_prediction = DTC.predict(X_test)
RF_prediction = RF.predict(X_test)
KNN_prediction = KNN.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
def evaluatePerformace(Prediction , actualValue, model):
    print(f"Prediction accuracy score for {model} => {accuracy_score(Prediction, actualValue)}\n")
    print(f"Confussion matrix for {model} '\n' {confusion_matrix(Prediction, actualValue)}\n")
    print(f"Classification report for {model}'\n' {classification_report(Prediction, actualValue)}")
```

```python
evaluatePerformace(DTC_prediction, y_test, 'Decision Tree')
```

```python
evaluatePerformace(RF_prediction, y_test, 'Random Forest')
```

```python
evaluatePerformace(KNN_prediction, y_test, 'K-Nearest Neighbors')
```

```python
evaluatePerformace(SVC_prediction, y_test, 'Support Vector Machine')
```

```python
accuracy_score = pd.DataFrame([
    ['Decision tree',accuracy_score(DTC_prediction, y_test)],
    ['K-Nearest Neighbors',accuracy_score(KNN_prediction, y_test)], 
    ['Support Vector Machine', accuracy_score(SVC_prediction, y_test)],
    ['Random Forest',accuracy_score(RF_prediction, y_test)]
],                  
columns = ['Model', 'Prediction Accuracy Score'])
```

```python
accuracy_score
```

```python
sns.set(rc={'figure.figsize':(15.0,5.5)})
sns.barplot(x='Model', y='Prediction Accuracy Score',data=accuracy_score, palette='husl')
plt.savefig('visuals/prediction_accuracy_emp_status.png')
```

From the plot above we can see the Random Forest algorithm came out with the best prediction accuracy score, followed by the Decision tree algorithm then K-Nearest Neighbors and finally Support Vector Machine
